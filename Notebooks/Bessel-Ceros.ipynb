## Ceros de las funciones de Bessel de orden entero $J_m(\alpha_{m,n})=0$:

En el módulo `scipy.special` las raíces de las funciones de Bessel están implementadas en la función `jn_zeros`. En particular `jn_zeros(m,N)` entrega un arreglo con las primeras $N$ raíces de la función $J_m$.

```python
for m in range(5):
    print('m = '+str(m)+': '+str(jn_zeros(m,5)))
```

Podemos graficar estas raíces, de la forma siguiente:

```python
figure()
nmax = 20
mmax = 10
for m in range(mmax):
    scatter(m*ones(nmax),jn_zeros(m,nmax))
ylim(0,50)
xlim(-1,mmax)
xlabel('$m$', fontsize=15)
ylabel(r'$\alpha_{m,n}$', fontsize=15)
xticks(range(mmax))
grid()
```

Note que estas raíces no están igualmente espaciadas, lo que se aprecia mejor al graficar la _diferencia entre una raiz y la siguiente_:

```python
figure()
nmax = 20
mmax = 10
for m in range(mmax):
    plot(range(nmax-1),diff(jn_zeros(m,nmax)), marker='o', markersize=5, lw=1, label='$m = %d$'%m)
xlim(-1,nmax-1)
ylim(3,)
xlabel('$n$', fontsize=15)
ylabel(r'$\alpha_{m,n+1}-\alpha_{m,n}$', fontsize=15)
xticks(range(nmax))
legend()
grid()
```

Vemos sin embargo que la diferencia $\alpha_{m,n+1}-\alpha_{m,n}$ tiende a $\pi$ para valores grandes de $n$.

## Ceros de la derivada de las funciones de Bessel de orden entero $J'_m(\alpha_{m,n})=0$:

Análogamente, las raíces de las derivadas $J'_m$ son calculadas por la función `jnp_zeros` de `scipy.special`:

```python
for m in range(5):
    print('m = '+str(m)+': '+str(jnp_zeros(m,5)))
```

```python
figure()
nmax = 20
mmax = 10
for m in range(mmax):
    scatter(m*ones(nmax),jnp_zeros(m,nmax))
ylim(0,50)
xlim(-1,mmax)
xlabel('$m$', fontsize=15)
ylabel(r'$\beta_{m,n}$', fontsize=15)
xticks(range(mmax))
grid()
```

En este caso, el gráfico de las diferencias es también similar al caso anterior:

```python
figure()
nmax = 20
mmax = 10
for m in range(mmax):
    plot(range(nmax-1),diff(jnp_zeros(m,nmax)), marker='o', markersize=5, lw=1, label='$m = %d$'%m)
xlim(-1,nmax-1)
ylim(3,)
xlabel('$n$', fontsize=15)
ylabel(r'$\beta_{m,n+1}-\beta_{m,n}$', fontsize=15)
xticks(range(nmax))
legend()
grid()
```
